In this problem, we slightly modify the AdaBoost algorithm to better explore some properties of the algorithm. Specifically, we no longer normalize the weights on the training examples after each iteration. The modified algorithm, which is set to run for $T$ iterations, is shown in Algorithm I.

Note that in the modified version, the weights associated with the training examples are no longer guaranteed to sum to one after each iteration (and therefore can not be viewed as a "distribution"), but the algorithm is still valid. Let us denote the sum of weights at the start of iteration $t$ by $Z_t = \sum_{i=1}^{n}w_i^{(t)}$. At the start of the first iteration of boosting, $Z_1 = n$. Let us now investigate the behavior of $Z_t$, as a function of t

**A:** At the $t^{th}$ iteration, we found a weak classifier that achieves a weighted training error $\epsilon_t$. Show that the choice, $\alpha_t = \frac{1}{2}\log\frac{1 - \epsilon_t}{\epsilon_t}$ is the optimal in the sense that it minimizes $Z_{t+1}$

*Hint: Look at $Z_{t+1}$ as function of $\alpha$ and find the value of $\alpha$ for which the function achieves the minimum. You may also find the following notational shorthand useful:

$$W_t = \sum_{i=1}^{n}w_i^{(t)}(1 - \delta(y_i, h_t(x_i)))$$
$$W_c = \sum_{i=1}^{n}w_i^{(t)}(\delta(y_i, h_t(x_i)))$$

where $W_c$ is the total weight of the points classified correctly by $h_t$ and $W_t$ is the total weight of the misclassified points. $\delta(y, h_t(x)) = 1$ whenever the label predicted by $h_t$ is correct and zero otherwise. The weights here are those available at the start of iteration $t$

**Solution:** <br/>
$y_i$ is the true label<br/>
$h_t(x_i)$ is the output of the weak classifier $h_t$<br/>
Since, it is a binary classification problem, for misclassified points, $y_ih_t(x_i)$ would be -1. And for correctly classified points, $y_ih_t(x_i)$ would be +1.<br/><br/>
Updated weight for misclassified points, $w_i^{(t+1)} = e^{\alpha_t}$  <br/> 
Total weight for misclassified points at $t^{th}$ iteration, <br/>
$W_t = \sum_{i=1}^{n}w_i^{(t)}(1 - \delta(y_i, h_t(x_i)))$ <br/>
$\implies W_t = \sum_{i=1}^{n}w_i^{(t)}$, since $\delta(y, h_t(x)) = 0$ <br/>
Total weight for misclassified points at $t+1^{th}$ iteration, <br/>
$W_{t+1} = \sum_{i=1}^{n}w_i^{(t+1)}$ <br/>
$\implies W_{t+1} = \sum_{i=1}^{n}w_i^{(t)}e^{\alpha_t}$ <br/>
$\implies W_{t+1} = W_te^{\alpha_t}$ <br/><br/>
Updated weight for correctly classified points, $w_i^{(t+1)} = e^{-\alpha_t}$ <br/>
Similarly, Total weight for correctly classified points at $t+1^{th}$ iteration, <br/>
$W_{c+1} = W_ce^{-\alpha_t}$ <br/><br/>
Sum of weights at $t+1^{th}$ iteration as a function of $\alpha$, <br/>
$Z_{t+1}(\alpha_t) = W_{t+1} + W_{c+1}$ <br/>
$\implies Z_{t+1}(\alpha_t) = W_te^{\alpha_t} + W_ce^{-\alpha_t}$ 

From the definition of weighted training error, we see that <br/>
$\epsilon_t = \frac{1}{Z_t}\sum_{i=1}^{n}w_i^{(t)}(1 - \delta(y_i, h_t(x_i)))$ <br/>
$\implies \epsilon_t = \frac{W_t}{Z_t}$ <br/>
$\implies W_t = Z_t\epsilon_t$ <br/>
Similary, $W_c = Z_t{(1-\epsilon_t)}$

To find the value for which $Z_{t+1}(\alpha_t)$ achieves the minimum, we take its derivative w.r.t $\alpha_t$ where it is 0.<br/>
$\frac{\partial Z_{t+1}}{\partial \alpha_t} = W_te^{\alpha_t} - W_ce^{-\alpha_t} = 0$<br/>
$\implies e^{2\alpha_t} = \frac{W_c}{W_t}$ <br/>
$\implies \alpha_t = \frac{1}{2}\log\frac{W_c}{W_t}$ <br/>

From the values of $W_t$ and $W_c$ from above, it is proved that <br/>
$\alpha_t = \frac{1}{2}\log\frac{1 - \epsilon_t}{\epsilon_t}$ is optimal.

**B:** Show that the sum of weights $Z_t$ is monotonically decreasing as a function of $t$.

**Solution:** <br/>
$Z_{t+1}(\alpha_t) = W_te^{\alpha_t} + W_ce^{-\alpha_t}$ 
At $\alpha_t = 0$, <br/>
$Z_{t+1}(0) = W_t + W_c = Z_t$<br/> 
Hence, $Z_{t+1} = \min_{\alpha_t} Z_{t+1} (\alpha_t) < Z_{t+1} (0) = Z_t$, when the minimizing $\alpha_t$ is non-zero. <br/>
$\alpha_t = \frac{1}{2}\log\frac{1 - \epsilon_t}{\epsilon_t}$ is non-zero when <br/>
$\log\frac{1 - \epsilon_t}{\epsilon_t} = 0 $<br/>
$\implies \frac{1 - \epsilon_t}{\epsilon_t} = 1 $<br/>
$\implies {\epsilon_t} = \frac{1}{2} $ i.e when error of the weak classifier is $\frac{1}{2}$ <br/>

To see how much $Z_t$ decreases as a function of t, <br/>
$Z_{t+1} = W_te^{-\frac{1}{2}\log\frac{W_c}{W_t}} + W_ce^{\frac{1}{2}\log\frac{W_c}{W_t}}$ <br/>
= $2\sqrt {W_tW_c}$ <br/>

$Z_t = W_t + W_c$, where $W_t < W_c$ <br/>
We can check if $Z_{t+1}$ is less than $Z_t$ <br/>
$Z_t^2 - Z_{t+1}^2 = W_t^2 + W_c^2 + 2W_tW_c - 4W_tW_c$ <br/>
= $W_t^2 + W_c^2 - 2W_tW_c$ <br/>
= $(W_c - W_t)^2 $ which is > 0, <br/>
which proves that $Z_t$ is monotonically decreasing as a function of $t$ .

## Bagging
Bootstrap aggregating, also called bagging, is a machine learning ensemble meta-algorithm designed to improve the stability and accuracy of machine learning algorithms used in statistical classification and regression. It also reduces variance and helps to avoid overfitting. Although it is usually applied to decision tree methods, it can be used with any type of method. Bagging is a special case of the model averaging approach.

Given a standard training set $D$ of size n, bagging generates $N$ new training sets $D_i$, roughly each of size n * ratio, by sampling from $D$ uniformly and with replacement. By sampling with replacement, some observations may be repeated in each $D_i$ The $N$ models are fitted using the above $N$ bootstraped samples and combined by averaging the output (for regression) or voting (for classification). 

-Source [Wiki](https://en.wikipedia.org/wiki/Bootstrap_aggregating)

**Solution:** <br/>
From the k-fold cross validation, we see that the maximum accuracy is when ratio = 0.8, N = 30, Max Depth = 15 and Candidate Splits = 150<br/>
For depth, the  accuracy does not increase significantly beyond a certain depth (10 to 15). Increasing the depth would only overfit the data, hence it is a bad idea to do so. <br/>
We can however claim that the increasing accuracies are a result of increasing the number of learners (N=25). Increasing the ratio also has some impact on the overall accuracy as higher sample size, makes the classifiers very generalizable. <br/>
Increasing the number of candidate splits has some effect on the classifier as it does more computations to find the best split vector. However, after a certain number, it would be fruitless to go on for a large iteration. <br/>
